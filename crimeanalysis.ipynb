# What is the overall trend in reported crimes against women across Top 5 over the years?

```python
df["Total_crime"]=df['RAPE']+df['KIDNAP & ABDUCTION']+df['DOWRY DEATHS']+df['CRUETLY BY HUSBAND OR HIS RELATIVES']+df['HUMAN TRAFFICKING']+df['ASSAULT ON MODESTY']+df['SEXUAL HARRASSMENT']+df['IMORTAL TRAFFICKING']
over_all=(df.groupby(df["STATES"])["Total_crime"].sum()).sort_values(ascending=False)
states_1=["Uttar Pradesh","West Bengal","Madhya Pradesh","Rajasthan","Maharashtra"]
plt.bar(states_1,over_all.head(5),label="Total No. of Crimes")
plt.xlabel("State-->")
plt.ylabel("Total No of Crimes-->")
plt.title("Total No of Crime in Top States and Union in India.")
plt.legend()
plt.xticks(rotation=35)
over_all.head(5)
```

The Total Number of Crimes in the Top 5 States in India from 2011 to 2021.

->Uttar Pradesh had the highest number of crimes, totaling 351,010 incidents.  
->West Bengal reported 266,132 crimes during the same period.  
->Madhya Pradesh recorded 263,377 crimes.  
->Rajasthan had 243,040 reported crimes.  
->Maharashtra witnessed 232,666 incidents of crime.

# How have reported crimes against women in India evolved over the years?
